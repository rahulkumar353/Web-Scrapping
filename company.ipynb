```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
headers={'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64;x64)AppleWebkit/537.36 (KHTML,like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage=requests.get('https://www.ambitionbox.com/list-of-companies?page-1',headers=headers).text
```

```python
soup=BeautifulSoup(webpage, 'lxml')
```

```python
company=soup.find_all('div' , class_='company-content-wrapper')
```

```python
len(company)
```

```python
name=[]
rating=[]
review=[]
c_sector=[]
headquarter=[]
c_old=[]
employees=[]

for i in company:
    name.append(i.find('h2').text.strip())
    rating.append(i.find('p', class_='rating').text.strip())
    review.append(i.find('a',class_='review-count').text.strip())
    c_sector.append(i.find_all('p',class_='infoEntity')[0].text.strip())
    headquarter.append(i.find_all('p',class_='infoEntity')[1].text.strip())
    c_old.append(i.find_all('p',class_='infoEntity')[2].text.strip())
    try:
        employees.append(i.find_all('p',class_='infoEntity')[3].text.strip())
    except:
        employees.append(np.nan)

d={'name':name,'ratings':rating,'reviews':review,'sector':c_sector,'headquarter':headquarter,'c_old':c_old,'employees':employees}
df=pd.DataFrame(d)
```

```python
#for multiple pages

final=pd.DataFrame()

for j in range(1,30):
        if(j==1):
             url='https://www.ambitionbox.com/list-of-companies?page-1'
        else:
             url='https://www.ambitionbox.com/list-of-companies?page-1&page={}'.format(j)
                
        headers={'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64;x64)AppleWebkit/537.36 (KHTML,like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
        webpage=requests.get(url,headers=headers).text
           
        soup=BeautifulSoup(webpage, 'lxml')
        company=soup.find_all('div', class_='company-content-wrapper')
        
        name=[]
        rating=[]
        review=[]
        c_sector=[]
        headquarter=[]
        c_old=[]
        employees=[]

        for i in company:
            name.append(i.find('h2').text.strip())
            try:
                rating.append(i.find('p', class_='rating').text.strip())
            except:
                rating.append(np.nan)
            try:
                review.append(i.find('a',class_='review-count').text.strip())
            except:
                review.append(np.nan)
            try:
                c_sector.append(i.find_all('p',class_='infoEntity')[0].text.strip())
            except:
                c_sector.append(np.nan)
            try:
                headquarter.append(i.find_all('p',class_='infoEntity')[1].text.strip())
            except:
                headquarter.append(np.nan)
            try:
                c_old.append(i.find_all('p',class_='infoEntity')[2].text.strip())
            except:
                c_old.append(np.nan)
            try:
                employees.append(i.find_all('p',class_='infoEntity')[3].text.strip())
            except:
                employees.append(np.nan)

        d={'name':name,'ratings':rating,'reviews':review,'sector':c_sector,'headquarter':headquarter,'c_old':c_old,'employees':employees}
        df=pd.DataFrame(d)
        
        final=final.append(df, ignore_index=True)
    
    
```
